```python
doc_list=doc.split(r'Receipt Number')
print(len(doc_list))
doc_list
```

```python
receipts_dict = {
    "date": [],
    "quantity": [],
    "total_amount": []
}
```

```python
for i in range(1,100):
    pattern=r'\d{2}.\d{2}.\d{4}'
    a=(re.findall(pattern,doc_list[i]))
    receipts_dict["date"].append(a[0])
```

```python
for i in range(1,100):
    pattern2=r'\d+\.\d{2}\s€'
    a=(re.findall(pattern2,doc_list[i]))
    receipts_dict["total_amount"].append(a[0])
```

```python
for i in range(1,99):
    pattern3=r'\s{4}\d+\d'
    a=(re.findall(pattern3,doc_list[i]))
    receipts_dict["quantity"].append(a[0])
receipts_dict["quantity"]
```
